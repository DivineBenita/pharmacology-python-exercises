```python
def dose_calculator(weight_kg):
    if weight_kg < 40:
        return 250
    elif weight_kg <= 70:
        return 500
    elif weight_kg <= 100:
        return 750
    else:
        return 1000

for w in [35, 55, 85, 110]:
    print(f"{w}kg -> {dose_calculator(w)}mg")
    
```

```python
def renal_adjustment(CrCl):
    if CrCl >= 60:
        return "Normal dose"
    elif CrCl >= 30:
        return "reduce dose by 50%"
    elif CrCl >= 15:
        return "reduce dose by 75%"
    else:
        return "Avoid drug - consider alternative"

for c in [75, 25, 20, 8]:
    print(f"CrCl{c}: {renal_adjustment(c)}")
```
