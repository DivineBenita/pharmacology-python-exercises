```python
def calculate_dose(weight_kg, drug = 'paracetamol', frequency = 3):
    if drug == 'paracetamol':
        dose_per_kg = 15
    elif drug == 'ibuprofen':
        dose_per_kg = 12
    else:
        dose_per_kg = 10
    single_dose = weight_kg * dose_per_kg
    daily_total = single_dose * frequency
    return single_dose, daily_total

single, daily = calculate_dose(60)
print(f"Single dose: {single}mg, daily_total: {daily}mg")
    
single, daily = calculate_dose(60, drug = 'paracetamol', frequency = 2)
print(f"Single dose: {single}mg, daily_total: {daily}mg")
```

```python
def cockcroft_gault(age, weight_kg, serum_creatinine, sex = 'male'):
    """Calculate estimated creatinine clearance in mL/min."""
    crcl = ((140 - age) * weight_kg / 72 * serum_creatinine) 
    if sex == 'female':
        crcl = crcl * 0.85
    return crcl

result1 = cockcroft_gault(45, 70, 1.0)
print(f"creatinine clearance: {result1:.1f}ml/min")

result2 = cockcroft_gault(65, 60, 1.2, sex = 'female')
print(f"creatinine clearance: {result2:.2f}ml/min")
```
