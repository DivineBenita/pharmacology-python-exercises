### Mean
- The mean is made the center of the curve
- If all data is used, the mean is not an estimate, but if a small sample is used to represent a large population, then it is an estimate.
- Estimated mean, also called sample mean, is different from mean.


### Variance 
- The population variance is gotten first by subtracting the population mean from each measurement, squaring each, summing up and dividing by the total sample size.
- When working with a small sample, the estimated population variance is calculated using this formula, only that 1 is subtracted from sample size used.

### Standard deviation
- The variance is the variables squared. Therefore, it cannot be plotted on the curve. The square root is gotten and this square root is known as the **standard deviation.**
- When writing, we write it as **mean+_ standard deviation**

#### Summary
With all of the data, we can calculate population mean, variance and standard deviation, but with limited data, we can calculate estimated mean, variance and standard deviation by tweaking the formulae slightly.
The mean, population or estimated is plotted plus or minus the standard deviation on the curve.
In real life, we always never have population data, **Microsoft Excel** calculates population variance using VAR.P while calculating estimated variance with VAR.S. Most, if not all of the time, we'd be using VAR.S.
