1. Mean: This is the average value of a given set of data.
2. Median: When the data is arranged from the smallest to the biggest, this is the middle value.
3. Mode: This is the most frequent data of the sample.
4. Range: This is the maximum minus the minimum data.
5. Variance: This is the average squared differences from the mean.
6. Standard Deviation: This is the square root of the variance. It tells you how far each variable is from the mean.
7. Normal Distribution: This is a perfectly symmetrical curve that shows the range of outcome of a given data.
8. Empirical Rule: This states that most outcomes will fall within mean plus or minus the standard deviation; 68% will fall within mean - 1SD to mean + 1SD. 95% will fall within mean - 2SD to mean + 2SD, 99.7% is beyond both.
9. Outlier: This is the value that is unusually higher or lower than the curve.
10. Skewness: This is when an outlier or abnormality leads to the curve not being perfectly symmetric but 'bent' to one side. Positively or negatively, depending on the side.
11. Population vs Sample: Population is the amount of people to be researched. A sample is a part of that data used to represent the entire population.
12. Correlation: This is when one variable increase or decrease proportionately with another variable. It doesn't imply causation.
