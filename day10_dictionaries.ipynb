## My bracket rule
Square brackets [ ] = position (first, second, last)
Curly braces { } = label (a named value)

```python
patient = {'name': 'Adaeze', 'age': 42, 'drug': 'metformin'}
print(patient)
```

```python
print(patient['name'])
print(patient['drug'])
```

```python
print(patient['weight_kg'])
```

```python
patient['weight_kg'] = 71.0
print(patient)
```

```python
patient['dose_mg'] = 1000
print(patient)
```

```python
if 'weight_kg' in patient: 
    print(f"Weight on file: {patient['weight_kg']}kg")
else: 
    print("Weight not recorded")
```

```python
del (patient['age'])
print(patient)
```
