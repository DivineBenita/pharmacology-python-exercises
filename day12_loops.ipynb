```python
patient_doses = [400, 750, 1200, 500, 1500, 300, 900]
safe_max = 1000

print("=== Dose Safety Checks ===")
for i, dose in enumerate(patient_doses):
    if dose > safe_max:
        print(f"Patient{i+1}: ALERT- {dose}mg exceeds limit")
    else: 
        print(f"Patient{i+1}: OK - {dose}mg")
```

```python
overdoses = [d for d in patient_doses if dose < safe_max]
print(f"Overdose count: {len(overdoses)}")
print(f"Overdose values: {overdoses}")
```
