```python
import pandas as pd

df = pd.read_csv('../data/diabetes.csv')

print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
```

```python
print(df.dtypes)
```

```python
print(df.describe())
```

```python
print(df['Outcome'].value_counts())
```

```python
diabetic = df[df['Outcome'] == 1]
non_diabetic = df[df['Outcome'] == 0]

print(f"Diabetic patients: {len(diabetic)}")
print(f"Non-Diabetic patients: {len(non_diabetic)}")
print(f"Mean glucose - Diabetic: {diabetic['Glucose'].mean():.1f}")
print(f"Mean glucose - Non-Diabetic: {non_diabetic['Glucose'].mean():.1f}")
```
