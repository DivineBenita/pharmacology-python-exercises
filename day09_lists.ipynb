```python
# A list of drug names — notice the square brackets
drugs = ['paracetamol', 'ibuprofen', 'aspirin', 'metformin', 'atorvastatin']

print(drugs)
print(type(drugs))
print(len(drugs))
```

```python
# Indexing — accessing one item by its position
print(drugs[0])   # first item
print(drugs[1])   # second item
print(drugs[4])   # fifth item
print(drugs[-1])  # last item
print(drugs[-2])  # second to last
```

```python
# Slicing — accessing a range of items
print(drugs[0:2])   # items at position 0 and 1 (not 2!)
print(drugs[1:4])   # items at position 1, 2, 3
print(drugs[:3])    # everything from the start up to position 3
print(drugs[2:])    # everything from position 2 to the end
```

```python
doses = [500, 400, 75, 850, 20]

# Add an item to the end
doses.append(1000)
print(doses)

# Remove a specific value
doses.remove(75)
print(doses)

# Change an item by its position
doses[0] = 600
print(doses)

# Sort the list
doses.sort()
print(doses)

# Sort in reverse
doses.sort(reverse=True)
print(doses)

# Check if a value exists in the list
print(500 in doses)
print(9999 in doses)
```

```python
# A ward of 8 patients — drug names in order
ward_drugs = ['metformin', 'lisinopril', 'atorvastatin', 
              'paracetamol', 'amoxicillin', 'metformin',
              'ibuprofen', 'lisinopril']

# Corresponding doses — same order, same position matches same patient
ward_doses = [850, 10, 40, 500, 500, 1000, 400, 5]

print(f"Number of patients:    {len(ward_drugs)}")
print(f"First Patient's Drug:  {ward_drugs[0]}")
print(f"Last Patient's Drug:   {ward_drugs[-1]}")

# Find the highest dose in the ward
highest_dose = max(ward_doses)
print(f"The Highest dose in the ward = {highest_dose}mg")

# Find the lowest dose in the ward
lowest_dose = min(ward_doses)
print(f"The Lowest dose in the ward = {lowest_dose}mg")

# Total of all doses
total_doses = sum(ward_doses)
print(f"The total of all doses = {total_doses}")

# Average of all doses~
average_dose = sum(ward_doses) / len(ward_doses)
print(f"The Average of all doses = {average_dose}")
```

```python
# A nested list — a list of lists
# Each inner list represents one patient: [name, age, drug, dose]
patient_records = [ ['Adaeze', 42, 'metformin', 850],
    ['Emeka', 67, 'lisinopril', 10],
    ['Fatima', 38, 'atorvastatin', 40], ]

print(patient_records[0])        # the entire first patient record
print(patient_records[0][0])     # just the name of the first patient
print(patient_records[1][2])     # just the drug of the second patient
print(patient_records[2][3])     # just the dose of the third patient
```
